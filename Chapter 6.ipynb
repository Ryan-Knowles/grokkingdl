# Chapter 6: Introduction to Backpropagation

## The streetlight problem

The problem: Given a traffic light with 3 signals, how do you know what each signal means? Which means walk and which means stop?


You are able to observe how other people behave with the signals.



So which signals are for walk? Which are for stop? Which don't matter?

This is how a neural network learns.

## Preparing the data
### Neural networks don't read streetlights

Supervised learning algorithms take one data set and turn it into another data set. This turns __what we know__ into __what we want to know__.



We need to identify the data for these two groups first before creating our neural network. Also, for some problems it is possible to go in reverse and flip the data sets, switching what is known and what we want to know.

## Matrices and the matrix relationship
### Translate the streetlights into math

Math doesn't understand streetlights. However, we can translate it into numbers by looking at the problem space:
1. There are 3 lights on a streetlight
2. Each light is on or off
3. Our desired knowledge is a binary option: walk or stop

So, we would go from an input space of 3 inputs with binary (on/off) values of 1 for on and 0 for off. And we can translate the output space into a single binary output of 1 for walk and 0 for stop.

The math would look something like this:$ \begin{bmatrix}0, 1, 1\end{bmatrix} \to 1$
